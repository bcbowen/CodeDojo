```python
players = {'Player 1': 'X', 'Player 2': 'O', 'Empty': '.'}

def init_board(): 
    board = []
    board.append(['.', '.', '.'])
    board.append(['.', '.', '.'])
    board.append(['.', '.', '.'])
    return board

def show_board(b):
    for row in b: 
        print(row)
            
def is_winner(board): 
    # check horizontal
    
    for i in range(0, 3):
        value = board[i][0]
        if value == players["Empty"]: 
            continue
        if board[i][1] == value and board[i][2] == value: 
            return (True, value, "Horizontal row {0}".format(i + 1))
    
    # check vertical 
    for i in range(0, 3):
        value = board[0][i]
        if value == players["Empty"]: 
            continue
        if board[1][i] == value and board[2][i] == value: 
            return (True, value, "Vertical col {0}".format(i + 1))
    

    # check diag starting with middle
    value = board[1][1]
    if value != players["Empty"]: 
        if board[0][0] == value and board[2][2] == value: 
            return (True, value, "Diag down")
        elif board[2][0] == value and board [0][2] == value: 
            return (True, value, "Diag up")
    return (False, players["Empty"], "No winner")

board = init_board()
```

```python
def run_game(): 
    show_about()
    board = init_board()
    activePlayer = 0; 
    show_board(board)
    gameOver = False
    while not gameOver:
        # [on_true] if [expression] else [on_false]
        playerKey = "Player 1" if activePlayer == 0 else "Player 2" #activePlayer == 0 then "Player 1" else "Player 2"
        player = players[playerKey]
        print("{0} is up".format(playerKey))
        activePlayer = 1 - activePlayer
               
        success = False
        while not success and not gameOver: 
            move = input("MOVE JERK\n")
            (success, board, message, gameOver) = process_move(move, board, player)
            if not success: 
                print(message)
                
        show_board(board)
        (isWinner, winner, message) = is_winner(board)
        if isWinner: 
            print(message)
            playAgain = input("Hit 'y' to play again").upper()
            if playAgain == "Y": 
                board = init_board()
            else: 
                print("GAME OVER")
                gameOver = True
        
        
        #gameOver = True
        
        
        
```

```python
def show_about(): 
    print("Tic Tac Toe")
    print("Players alternate between X and O, X goes first")
    print("Each player enters their next move in the form of Y, X, where Y is the row and X is the column")
    
def process_move(move, board, symbol):
    message = ""
    success = False
    gameOver = False
    if move.upper() == 'Q': 
        gameOver = True
        message = "The user has quit the game"
    else: 
        (success, message, y, x) = parse_symbols(move)
        if success: 
            if board[y][x] != '.': 
                success = False 
                message = "That move is unavailable"
            else: 
                message = "Your move has been made"
                board[y][x] = symbol
    return (success, board, message, gameOver)

def parse_symbols(input): 
    symbols = input.split(',')
    success = False
    message = "" 
    y = -1
    x = -1
    
    if len(symbols) != 2: 
        message = "Invalid input: '{0}'".format(input)
    else: 
        symbols[0] = symbols[0].strip()
        symbols[1] = symbols[1].strip()
        
        if not symbols[0].isdigit():
            message = "Invalid row: {0}".format(symbols[0])
        elif not symbols[1].isdigit(): 
            message = "Invalid col: {0}".format(symbols[1])
        else: 
            y = int(symbols[0])
            x = int(symbols[1])
            if y < 0 or y > 2: 
                message = "Row out of range: {0}".format(y)
            elif x < 0 or x > 2: 
                message = "Col out of range: {0}".format(x)
            else: 
                success = True
    return (success, message, y, x)
```

```python
def run_tests(): 
    board_displays()
    winner_tests()
    parse_symbols_tests()

def board_displays(): 
    print('Showing board:')
    board = init_board()
    show_board(board)
    
def run_winner_test(board, expected): 
    print("Testing Board")
    show_board(board)
    (expectedResult, expectedWinner, expectedMsg) = expected
    (result, winner, msg) = is_winner(board)
    print("result: {0} {1} {2}".format(result, winner, msg))
    fail = False
    if result != expectedResult: 
        print("Unexpected result: {0}".format(result))
        fail = True
    if winner != expectedWinner: 
        print("Unexpected winner: {0}".format(winner))
        fail = True
    if msg != expectedMsg: 
        print("Unexpected msg: {0}".format(msg))
        fail = True
    
    if not fail: 
        print("PASS")
    
def winner_tests(): 
    cases = [([['.', '.', '.'], ['.', '.', '.'], ['.', '.', '.']], (False, players["Empty"], "No winner")), 
             ([['X', '.', '.'], ['.', '.', '.'], ['.', '.', '.']], (False, players["Empty"], "No winner")), 
             ([['.', '.', '.'], ['.', 'X', '.'], ['.', '.', '.']], (False, players["Empty"], "No winner")), 
             ([['.', '.', '.'], ['.', '.', '.'], ['.', '.', 'X']], (False, players["Empty"], "No winner")), 
             ([['.', '.', 'O'], ['.', '.', '.'], ['.', '.', '.']], (False, players["Empty"], "No winner")), 
             ([['.', '.', '.'], ['.', 'O', '.'], ['.', '.', '.']], (False, players["Empty"], "No winner")), 
             ([['.', '.', '.'], ['.', '.', '.'], ['O', '.', '.']], (False, players["Empty"], "No winner")), 
        ([['X', 'X', 'X'], ['.', '.', '.'], ['.', '.', '.']], (True, 'X', "Horizontal row 1")),
        ([['.', '.', '.'], ['X', 'X', 'X'], ['.', '.', '.']], (True, 'X', "Horizontal row 2")),
        ([['.', '.', '.'], ['.', '.', '.'], ['X', 'X', 'X']], (True, 'X', "Horizontal row 3")),
        ([['O', '.', '.'], ['O', '.', '.'], ['O', '.', '.']], (True, 'O', "Vertical col 1")),
        ([['.', 'O', '.'], ['.', 'O', '.'], ['.', 'O', '.']], (True, 'O', "Vertical col 2")),
        ([['.', '.', 'O'], ['.', '.', 'O'], ['.', '.', 'O']], (True, 'O', "Vertical col 3")),
        ([['X', 'X', 'O'], ['.', '.', '.'], ['.', '.', '.']], (False, players["Empty"], "No winner")),
        ([['.', 'O', '.'], ['.', 'X', '.'], ['.', 'O', '.']], (False, players["Empty"], "No winner")),
        ([['X', '.', '.'], ['.', 'X', '.'], ['.', '.', 'X']], (True, 'X', "Diag down")),
        ([['.', '.', 'O'], ['.', 'O', '.'], ['O', '.', '.']], (True, 'O', "Diag up")),             
    ]
    
    for (board, expected) in cases: 
        run_winner_test(board, expected)
        
def parse_symbols_tests(): 
    cases = [("1, 2", True, "", 1, 2),
             ("1,2", True, "", 1, 2),
             ("yadda", False, "Invalid input: 'yadda'", -1, -1),
             ("1", False, "Invalid input: '1'", -1, -1),
             ("", False, "Invalid input: ''", -1, -1),
             ("a, 1", False, "Invalid row: a", -1, -1),
             ("1, a", False, "Invalid col: a", -1, -1),
             ("4, 1", False, "Row out of range: 4", 4, 1),
             ("1, 4", False, "Col out of range: 4", 1, 4),
    ]
    print("Parse Symbols Tests")
    test = 1
    for (value, expectedSuccess, expectedMessage, expectedY, expectedX) in cases: 
        (testPass, message) = run_symbol_test(value, expectedSuccess, expectedMessage, expectedY, expectedX)
        print("Test {0} {1} {2}".format(test, testPass, message))
        test += 1
    
def run_symbol_test(value, expectedSuccess, expectedMessage, expectedY, expectedX):
    testPass = True
    message = ""
    (success, message, y, x) = parse_symbols(value)
    if expectedSuccess != success: 
        testPass = False
        message += "Success Fail. Expected {0}, got {1}. ".format(expectedSuccess, success)
    if expectedMessage != message: 
        testPass = False
        message += "Message Fail. Expected {0}, got {1}. ".format(expectedMessage, message)
    if expectedY != y: 
        testPass = False
        message = "Unexpected Y! Expected {0}, got {1}. ".format(expectedY, y)
    if expectedX != x: 
        testPass = False
        message = "Unexpected X. Expected {0}, got {1}. ".format(expectedX, x)
        
    return (testPass, message)
    
```
