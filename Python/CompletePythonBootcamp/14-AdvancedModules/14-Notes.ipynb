### Regex

```python
import re
text = "my phone is a very shitty phone, dude"
match = re.search('phone', text)
print(match.span()) # only shows first match
print(match.start())
print(match.end())
```

```python
matches = re.findall('phone', text)
print(matches)
print(len(matches))
```

```python
# to loop through each match, use finditer
for match in re.finditer('phone', text): 
    print(match)
    print(match.group())
```

Capture with Compile:

```python
text2 = "Jenny's number is 305-867-5309"
phone_pattern = re.compile(r"(\d{3})-(\d{3})-(\d{4})")
results = re.search(phone_pattern, text2)
print(results.group())
# group ordering is 1-based
print(results.group(1))
print(results.group(2))
print(results.group(3))
```

Remove characters from string

```python
test_phrase = "We only want letters! But there is punctuation!! What can we do?"
pattern = r"[^!.? ]+"
cleaned = re.findall(pattern, test_phrase)
print(' '.join(cleaned))
```
